```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('datasets/pinglecourses.csv', delimiter = ',')
```

```python
df = df[~df['SchoolName'].isin(['SchoolName'])]
df = df.replace(to_replace= r'\r', value= '', regex=True)
df = df.replace({pd.np.nan: None})
```
